```python
import random

### SENTENCE GENERATOR: UPON USER REQUEST FOR A NICE OR A MEAN COMMENT, PROGRAM WILL RANDOMLY CHOOSE WORDS OF CORRESPONDING MOOD TO CREATE A SENTENCE (WHICH IDEALLY WILL HAVE MEANING :) ) 

# Lists that will contains instances created for each word type. That is where all of our words are stored. I could split them directly to have a list for each category and each mood, but then allows to repurpose the database for othe program ideas.
nouns = []
verbs = []
adjectives = []
adverbs = []

# Lists where different word types that match the requested mood will be stored (program will pick at random through those to genereate a sentence)
n_possibilities = []
vb_possibilities = []
adj_possibilities = []
adv_possibilities = []

#DIFFERENT CLASSES OF WORD CATEGORIES
     
class ArticleAndNoun:
  
  def __init__(self, word, category, mood, number):
    nouns.append(self)
    self.word = word
    self.category = category
    self.mood = mood
    self.number = number

  def get_category(self):
    return self.category
  def get_mood(self):
    return self.mood
  def get_number(self):
    return self.number

  def set_category(self, category):
    self.category = category
  def set_mood(self, mood):
    self.mood = mood
    
  def give_description(self):
      return "The word '" + self.word + "' is a " + self.number + " " +   self.category + " with a " + self.mood + " connotation."



class Verb:
  
  def __init__(self, word, category, mood, first_person, third_person):
    verbs.append(self)
    self.word = word
    self.category = category
    self.mood = mood
    self.first_person = first_person
    self.third_person = third_person

  def get_category(self):
    return self.category
  def get_mood(self):
    return self.mood
  def get_first_person(self):
    return self.first_person
  def get_third_person(self):
    return self.third_person
  
  
  def set_category(self, category):
    self.category = category
  def set_mood(self, mood):
    self.mood = mood
  def set_first_person(self, first_person):
    self.first_person = first_person
  def set_third_person(self, third_person):
    self.third_person = third_person

  def give_description(self):
      return "The word '" + self.word + "' is a " +   self.category + " with a " + self.mood + " connotation."

  
    
class Adjective:

  def __init__(self, word, category, mood):
    adjectives.append(self)
    self.word = word
    self.category = category
    self.mood = mood
    
  def get_category(self):
    return self.category
  def get_mood(self):
    return self.mood

  def set_category(self, category):
    self.category = category
  def set_mood(self, mood):
    self.mood = mood

  def give_description(self):
      return "The word '" + self.word + "' is an " +   self.category + " with a " + self.mood + " connotation."



class Adverb:

  def __init__(self, word, category, mood):   
    adverbs.append(self)
    self.word = word
    self.category = category
    self.mood = mood
    
  def get_category(self):
    return self.category
  def get_mood(self):
    return self.mood

  def set_category(self, category):
    self.category = category
  def set_mood(self, mood):
    self.mood = mood

  def give_description(self):
      return "The word '" + self.word + "' is an " +   self.category + " with a " + self.mood + " connotation."
    

  
# INSTANCEs OF WORDS OF EACH TYPE AND MOOD TO TRY OUT CODE
the_cat = ArticleAndNoun("the cat", "noun", "neutral", "singular")
a_jerk = ArticleAndNoun("a jerk", "noun", "negative", "singular")
a_lifesaver = ArticleAndNoun("a lifesaver", "noun", "positive", "singular")
dogs = ArticleAndNoun("dogs", "noun", "general", "plural")

have = Verb("have", "verb", "neutral", "have", "has")
be = Verb("are", "verb", "neutral", "am", "is")

amazingly = Adverb("amazingly", "adverb", "positive")
awfully = Adverb("awfully", "adverb", "negative")

beautiful = Adjective("beautiful", "adjective", "positive")
lazy = Adjective("lazy", "adjective", "negative")


#INPUT REQUEST:

#Now asking user to decide between nice and mean statement. If nice for instance, loops iterate through all lists of word categories and store positive and neutral moods into new lists. It will then pick one of each type at random and create a sentence.
#TODO1 will need to factor not more than one use of neutral.
#TODO2 could also split article and noun to insert adjective.
#TODO3 should also factor different types of sentences depending on verbs used (like he has + noun, or he is + adejctive or noun)
#TODO4 will need to make sure no repeat within the same sentence


class SentenceBuilder:
  def __init__(self, response):   
    self.response = response
    
    
response = str(input("Do you want me to say something 'nice' or 'mean'? "))
if response.lower() == 'nice':
  for noun in nouns:
    if noun.mood == "positive" or noun.mood == "neutral":
      n_possibilities.append(noun)
  for verb in verbs:
    if verb.mood == "positive" or verb.mood == "neutral":
      vb_possibilities.append(verb)
  for adj in adjectives:
    if adj.mood == "positive" or adj.mood == "neutral":
      adj_possibilities.append(adj)
  for adv in adverbs:
    if (adv.mood == "positive") or (adv == "neutral"):
      adv_possibilities.append(adv)

if response.lower() == 'mean':
  for noun in nouns:
    if noun.mood == "negative" or noun.mood == "neutral":
      n_possibilities.append(noun)
  for verb in verbs:
    if verb.mood == "negative" or verb.mood == "neutral":
      vb_possibilities.append(verb)
  for adj in adjectives:
    if adj.mood == "negative" or adj.mood == "neutral":
      adj_possibilities.append(adj)
  for adv in adverbs:
    if (adv.mood == "negative") or (adv == "neutral"):
      adv_possibilities.append(adv)

generated_sentence = random.choice(n_possibilities).word + " " + random.choice(adv_possibilities).word + " " + random.choice(vb_possibilities).third_person + " " + random.choice(n_possibilities).word + "."
print(generated_sentence)
```
